```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[6547])
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train[68]
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob=model.predict(x_test)
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

training_loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(training_loss) + 1)

fig, ax = plt.subplots(figsize=(8, 5))
ax.set_xlim(1, len(training_loss))
ax.set_ylim(min(min(training_loss), min(val_loss)) - 0.05, max(max(training_loss), max(val_loss)) + 0.05)
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.set_title('Training vs Validation Loss (Animated)')

line1, = ax.plot([], [], 'r-', label='Training Loss')
line2, = ax.plot([], [], 'b-', label='Validation Loss')
ax.legend()

def update(frame):
    line1.set_data(list(epochs[:frame]), training_loss[:frame])
    line2.set_data(list(epochs[:frame]), val_loss[:frame])
    return line1, line2

ani = animation.FuncAnimation(fig, update, frames=len(training_loss)+1, blit=True, interval=200)

HTML(ani.to_jshtml())
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

training_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
epochs = range(1, len(training_acc) + 1)


fig, ax = plt.subplots(figsize=(8, 5))
ax.set_xlim(1, len(training_acc))
ax.set_ylim(min(min(training_acc), min(val_acc)) - 0.02, 1.01)
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax.set_title('Training vs Validation Accuracy (Animated)')

line1, = ax.plot([], [], 'g-', label='Training Accuracy')
line2, = ax.plot([], [], 'm-', label='Validation Accuracy')
ax.legend(loc='lower right')

def update(frame):
    line1.set_data(list(epochs[:frame]), training_acc[:frame])
    line2.set_data(list(epochs[:frame]), val_acc[:frame])
    return line1, line2

ani_acc = animation.FuncAnimation(fig, update, frames=len(training_acc)+1, blit=True, interval=200)

HTML(ani_acc.to_jshtml())
```
